```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

<div style="text-align: center;">
<font face="Courier New" color="#f2f068" size="4"><b><u>
    PLOT

```python
x=[1,2,3,4,5]
y=[41,58,82,101,112]

# Size the graph. It should be at top. dpi: pixer per inch, dpi'i yüksek tutmaya çalış, zoomlayınca düzgün görülmesi için. 300 iyi ama save ederken de belirtebilirsin.
plt.figure(figsize=(6,5), dpi=100)

plt.plot(x,y, 'yo--', label='2x', 
    #    color='y', marker='o', linestyle='--',
         linewidth=1, markersize=5, markeredgecolor='black')

#Eğer diğer çizeceğini plt.show dedikten sonraya koyarsan ayrı grafik çizer:
x2=np.arange(1,5.5,.5) #5.5 hariç
# Bir değere kadar çizgili, sonrasını kesikli yapmak için: 
plt.plot(x2[:6], x2[:6]**3, 'rs-', label='x^2',
         linewidth=1, markersize=5, markeredgecolor='black') 
plt.plot(x2[5:], x2[5:]**3, 'rs--', # label'i tekrar belirtmeye gerek yok
         linewidth=1, markersize=5, markeredgecolor='black') 

plt.title('My Graph',fontdict={'fontname':'Comic Sans MS', 'fontsize':14})
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.xticks(np.arange(0,6,1)) #plt.xticks([0,1,2,3,4,5])
plt.yticks(np.arange(0,150,25)) # plt.yticks([0,25,50,75,100,125])
plt.legend() # label'ı plt.plot'ta belirtmelisin.
plt.savefig('mygraph.png', dpi=300) # saves to current directory. 300 dpi high res için iyi

plt.show() # to get rid of text above plot
```

```python
fifa=pd.read_csv('fifa_data.csv')
fifa
```

```python
bins=[40,50,60,70,80,90,100] # bins histogramda kullanılıyor. her bir barın uzunluğu
plt.hist(fifa.Overall, bins=bins)

plt.xticks(bins)
plt.xlabel('Skill Level')
plt.ylabel('Number of Players')

plt.show()
```

<div style="text-align: center;">
<font face="Courier New" color="#f2f068" size="4"><b><u>
    PIE CHART

```python
left=fifa.loc[fifa['Preferred Foot']=='Left'].count()[0]
right=fifa.loc[fifa['Preferred Foot']=='Right'].count()[0]
labels=['Left','Right']
colors=['red','yellow']
plt.pie([left,right],labels=labels,colors=colors,autopct='%.2f')
```

```python
fifa.Weight=[int(x.strip('lbs')) if type(x)==str else x for x in fifa.Weight]
fifa.Weight
```

```python
light=fifa.loc[fifa.Weight<150].count()[0]
medium=fifa.loc[(fifa.Weight>=150) & (fifa.Weight<175)].count()[0]
heavy=fifa.loc[fifa.Weight>175].count()[0]

weights=[light,medium,heavy]
labels=['<150','150-175','>175']
explode=(.1,0,0)
plt.style.use('ggplot')

plt.pie(weights, labels=labels, explode=explode, autopct='%.2f') # yüzdeler çok yakın yazıldığında pctdistance=0.8 ekleyebilirsin.
plt.show()
```
